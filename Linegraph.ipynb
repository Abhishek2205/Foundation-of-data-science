```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([2,8,6,5,7,9])
y=np.array([3,8,1,10,5,7])
plt.figure(figsize=(10,5))
plt.plot(x,y,marker='o')
for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1, 2, 3, 4])
y = np.array([5,10,15,20])

# first plot with X and Y data
plt.plot(x, y)

y1 = [6,12,18,24]

# second plot with x and y1 data
plt.plot(x, y1, '-.')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
X = np.linspace(0, 3, 200)
Y1 = X**2 + 3
Y2 = np.sin(X)

plt.plot(X, Y1, '-',lw=4)
plt.plot(X, Y2, '-.',lw=4)

plt.fill_between(X, Y1,Y2, color='red', alpha=.5)

plt.show()
```
